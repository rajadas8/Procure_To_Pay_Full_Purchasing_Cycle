## Procure-to-Pay(P2P) - Full Purchasing Cycle (SAP MM)

### Step 1: Import Required Libraries

```python
import pandas as pd
from datetime import datetime
import uuid
```

### Step 2: Create Master Data (Vendor, Material, Company)

```python
# Vendor Master
vendors = pd.DataFrame({
    "Vendor_ID": [101, 102, 103],
    "Vendor_Name": ["ABC Suppliers", "XYZ Traders", "Global Tech"],
    "City": ["Kolkata", "Delhi", "Mumbai"]
})

# Material Master
materials = pd.DataFrame({
    "Material_ID": [1001, 1002, 1003],
    "Material_Name": ["Laptop", "Mouse", "Keyboard"],
    "Price": [50000, 500, 1500],
    "Stock": [10, 50, 30]
})

# Company Data
company = {
    "Company_Code": "C001",
    "Company_Name": "Tech Solutions Ltd"
}

print("Master Data Loaded")
vendors, materials
```

### Step 3: Create Purchase Requisition (PR)

```python
def create_pr(material_id, quantity):
    pr = {
        "PR_ID": str(uuid.uuid4())[:8],
        "Material_ID": material_id,
        "Quantity": quantity,
        "Date": datetime.now(),
        "Status": "CREATED"
    }
    return pr

pr1 = create_pr(1001, 2)
print("Purchase Requisition Created ")
pr1
```

### Step 4: Approve Purchase Requisition

```python
def approve_pr(pr):
    pr["Status"] = "APPROVED"
    pr["Approved_Date"] = datetime.now()
    return pr

pr1 = approve_pr(pr1)
print("PR Approved")
pr1
```

### Step 5: Request for Quotation (RFQ) & Vendor Selection

```python
def select_vendor(material_id):
    # Simple logic: choose first vendor
    return vendors.iloc[0]["Vendor_ID"]

selected_vendor = select_vendor(pr1["Material_ID"])
print("Vendor Selected:", selected_vendor)
```

### Step 6: Create Purchase Order (PO)

```python
def create_po(pr, vendor_id):
    po = {
        "PO_ID": str(uuid.uuid4())[:8],
        "Vendor_ID": vendor_id,
        "Material_ID": pr["Material_ID"],
        "Quantity": pr["Quantity"],
        "PO_Date": datetime.now(),
        "Status": "CREATED"
    }
    return po

po1 = create_po(pr1, selected_vendor)
print("Purchase Order Created")
po1
```

### Step 7: Approve Purchase Order

```python
def approve_po(po):
    po["Status"] = "APPROVED"
    po["Approved_Date"] = datetime.now()
    return po

po1 = approve_po(po1)
print("PO Approved ")
po1
```

### Step 8: Goods Receipt (GR)

```python
def goods_receipt(po):
    gr = {
        "GR_ID": str(uuid.uuid4())[:8],
        "PO_ID": po["PO_ID"],
        "Material_ID": po["Material_ID"],
        "Quantity": po["Quantity"],
        "GR_Date": datetime.now(),
        "Status": "RECEIVED"
    }

    # Update stock
    materials.loc[materials["Material_ID"] == po["Material_ID"], "Stock"] += po["Quantity"]

    return gr

gr1 = goods_receipt(po1)
print("Goods Received ")
gr1
```

### Step 9: Invoice Verification (IV)

```python
def create_invoice(gr):
    material = materials[materials["Material_ID"] == gr["Material_ID"]].iloc[0]

    total_amount = material["Price"] * gr["Quantity"]

    invoice = {
        "Invoice_ID": str(uuid.uuid4())[:8],
        "GR_ID": gr["GR_ID"],
        "Vendor_ID": po1["Vendor_ID"],
        "Amount": total_amount,
        "Status": "VERIFIED",
        "Invoice_Date": datetime.now()
    }

    return invoice

invoice1 = create_invoice(gr1)
print("Invoice Created & Verified")
invoice1
```

### Step 10: Payment Processing

```python
def process_payment(invoice):
    payment = {
        "Payment_ID": str(uuid.uuid4())[:8],
        "Invoice_ID": invoice["Invoice_ID"],
        "Amount": invoice["Amount"],
        "Payment_Date": datetime.now(),
        "Status": "PAID"
    }
    return payment

payment1 = process_payment(invoice1)
print("Payment Done")
payment1
```

### Step 11: Full P2P Automation Function

```python
def full_p2p_cycle(material_id, quantity):
    print("\n Starting Full P2P Cycle...\n")

    pr = create_pr(material_id, quantity)
    pr = approve_pr(pr)

    vendor = select_vendor(material_id)

    po = create_po(pr, vendor)
    po = approve_po(po)

    gr = goods_receipt(po)

    invoice = create_invoice(gr)

    payment = process_payment(invoice)

    print(" P2P Cycle Completed Successfully!\n")

    return {
        "PR": pr,
        "PO": po,
        "GR": gr,
        "Invoice": invoice,
        "Payment": payment
    }

result = full_p2p_cycle(1002, 5)
result
```

### Step 12: Reporting Dashboard

```python
def generate_report(pr, po, gr, invoice, payment):
    report = pd.DataFrame({
        "Stage": ["PR", "PO", "GR", "Invoice", "Payment"],
        "Status": [
            pr["Status"],
            po["Status"],
            gr["Status"],
            invoice["Status"],
            payment["Status"]
        ]
    })
    return report

report = generate_report(pr1, po1, gr1, invoice1, payment1)
print("\n P2P Report:")
report
```
